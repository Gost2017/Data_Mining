```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

plt.grid(color='g', linestyle='--', linewidth=0.5)
plt.title('Start new plot')
plt.xlabel('X lable')
plt.ylabel('Y lable')

plt.show()
```

### Line Plot

```python
x = np.arange(0,10,.1)

plt.figure(figsize = (7,7))

plt.plot(x,np.sin(x),'-',label='Sin',c = 'g')
plt.plot(x,np.cos(x),'--',label='Cos')

plt.grid(color='g', linestyle='--', linewidth=0.5)
plt.title('Line plot')
plt.xlabel('X lable')
plt.ylabel('Y lable')
plt.legend()

plt.show()
```

#### Scatter Plot

```python
x = np.arange(0,10,.1)

plt.figure(figsize = (7,7))
color = np.array([1,2,3,4,5])

plt.scatter(x,np.sin(x),c='b',marker = 's' ,edgecolor ="black",label='Sin', alpha = .25, s=x*3)
plt.scatter(x,np.cos(x),c='g',marker = '^' ,edgecolor ="red",label='Cos', alpha = .25, s=x*3)

plt.grid(color='g', linestyle='--', linewidth=0.5)
plt.title('Scatter Plot new plot')
plt.xlabel('X lable')
plt.ylabel('Y lable')
plt.legend()

plt.show()
```

#### Bar Chart

```python
line_data = np.array([1,2,3,4,5,6])
xval = range(len(line_data))

plt.bar(xval,line_data, width = 0.3)
plt.title(xval)

plt.grid(color='g', linestyle='--', linewidth=0.5)
plt.title('Bar Chart')
plt.xlabel('X lable')
plt.ylabel('Y lable')

plt.show()
```

#### Subplots

```python
x = np.arange(0,10,.1)

plt.figure(figsize = (7,7))

plt.subplot(1,3,1)
plt.plot(x,np.sin(x),'-',label='Sin')
plt.plot(x,np.cos(x),'--',label='Cos')

plt.subplot(132)
plt.scatter(x,np.sin(x),c='b',marker = 's' ,edgecolor ="black",label='Sin', alpha = .25, s=x*3)
plt.scatter(x,np.cos(x),c='g',marker = '^' ,edgecolor ="red",label='Cos', alpha = .25, s=x*3)

plt.subplot(133)
plt.bar(xval,line_data, width = 0.3)

plt.show()
```

```python
x = np.arange(0,10,.1)

plt.figure(figsize = (7,7))

fig, axe = plt.subplots(1,3,figsize=(10,3))

axe[0].plot(x,np.sin(x),'-',label='Sin')
axe[0].plot(x,np.cos(x),'--',label='Cos')
axe[0].set_title('Lineplot')

axe[1].scatter(x,np.sin(x),c='b',marker = 's' ,edgecolor ="black",label='Sin', alpha = .25, s=x*3)
axe[1].scatter(x,np.cos(x),c='g',marker = '^' ,edgecolor ="red",label='Cos', alpha = .25, s=x*3)
axe[1].set_title('Scatter plot')

axe[2].bar(xval,line_data, width = 0.3)
axe[2].set_title('Bar Chart')


fig.suptitle('3 Plot in 1 row and 3 columns')

plt.show()
```

#### Histograms

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)
axs = [ax1,ax2,ax3,ax4]

# draw n = 10, 100, 1000, and 10000 samples from the normal distribution and plot corresponding histograms
for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    
    axs[n].hist(sample, bins = 100)
    axs[n].set_title('n={}'.format(sample_size))
```

#### Boxplot

```python
import pandas as pd
normal_sample = np.random.normal(loc=0.0, scale=1.0, size=10000)
random_sample = np.random.random(size=10000)
gamma_sample = np.random.gamma(2, size=10000)

df = pd.DataFrame({'normal': normal_sample, 
                   'random': random_sample, 
                   'gamma': gamma_sample})
df.describe()
```

```python
plt.figure()
plt.boxplot(normal_sample)
plt.show()
```

```python
plt.figure()
plt.boxplot([normal_sample,random_sample,gamma_sample], vert=False)
plt.show()
```

```python
plt.figure()
fig, axs = plt.subplots(2,1, sharex=True)

axs[0].hist(gamma_sample, bins=1000)
axs[1].boxplot(gamma_sample, vert = False)

plt.show()
```

```python
import scipy.stats as stats
density = stats.gaussian_kde(gamma_sample)

plt.figure()
n, x, _ = plt.hist(gamma_sample, bins=100, density=True)

plt.plot(x, density(x))

plt.show()
```
